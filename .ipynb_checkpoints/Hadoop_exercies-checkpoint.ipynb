**A.** Hadoop has a replication factor of three(by default) in which each block gets replicated in case of any node failure (faults) can be handled

**Q. What is the safe mode in Hadoop**
** **
**A.** HDFS safe mode in Hadoop is also a maintenance mode. It is a read-only mode that restrict any modification to the file system.
- When Namenode gets restarted. To create a FS_image (to serve client) namenode goes to **safe mode** and do checkpointing (merge edit log and fs_image) and leave safe mode. (namenode is unserviceable that time).
- **Caster initialization** : when an HDFS cluster starts, the name node enters safe-mode to ensure that all data-nodes are up, and running and all blocks are reported
- **Manually safe mode** : if required you can use safe mode manually there are some commands     
    `> hdfs dfsadmin -safemode enter`          
    `> hdfs dfsadmin -safemode get`            
    `> hdfs dfsadmin -safemode leave`

**Q. how long heart-beat interval**
** **
**A.** Heartbeat is a signal from a datanode to namenode in Hadoop 
- 3s alive
- 10s stale
- 10 minutes dead

**Important point related to heartbeat**
- Heartbeat single carries data node storage information liveness confirmation and Block reports.
- Namenode does not store data on estabdata node (heartbeat Signal 10 second)

**Q. Explain HDFS architecture**
****
**A.** 
- **NameNode** : NameNode store metadata about file, directory and block location. **The namespace(namenode) tracks file blocks but neighbor put actual data content**
- **DataNode** : DataNodes are workdernode, the actual data blocks of the file is stored datanode, communicates a regularly with the namenode, and reporting the blocks it holds and it health status.


![HDFS-Architecture1.png](images/HDFS-Architecture1.png)

**Important points in HDFS**             

- HDFS can store blocks on different data node and also different rack
- Data nodes can communicate with each other for various purposes data application, block management, and health monitoring
- When a client obtaining permission from the namenode it can communicate directly with data node to perform read and write operations
- Metadata is initially stored on the namenode hard disk is in operation, this metadata loaded into the name nodes RAM for faster access and efficient management
- When to put a small file on HDFS it will occupy a full block-space

**Q. What is the speculative execution in Hadoop**
** **
**A.** In Hadoop during speculative execution a certain number of duplicate tasks are launched on a different slave node, and multiple copies of the same map or reduce task can be executed using **speculative execution**.


**Q. What is the high availability in Hadoop**
** **
**A.** The high availability feature in Hadoop ensure the availability of the Hadoop cluster without any downtime even in unfavorable condition like namenode failure, datanode failure, machine crash, etc. It means if the machine crashes data will be accessible from another path.

- High availability is the solution of a single point of failure
- High capability must be configured it is not a by default feature
- Zookeeper will decide active and stand by Namenode
- High availability used communication of active and stand by name node node
- Active name node use in_use.log file






**Q. what is the federation in Hadoop**
** **
**A.** **Vertical scaling problem at RAM of namenode** 
- HDFS federation improves the existing HDFS architecture through a clear separation of namespace(namenode) and storage, enabling generic blog storage layer. It enables support for multiple namespace(namenode) in the cluster to improve accessibility and isolation.
- In other words Federation in Hadoop is architecture that allows a single Hadoop cluster to scale by using multiple independent namenode


**Important points in federation**
- Federation means it is separation
- Data notes share metadata with the name node and Hadoop HDFS helps distribute it to different namenode to protect against name node failure
- You cannot access namenode metadata directly but must use the Associated Namenode for access

**Q. what is the reck awareness in Hadoop**
** **
**A.** Rack awareness in Hadoop is a concept to choice a nearby data node (closest to the client which has raised the read/write request), thereby reducing the network traffic. Hadoop supports the configuration of rack awareness to ensure the placement of one replica of the data block on a different rack.


**Note** 

- **Intra-Rack communication** : Involves notes within the same rack, and it is typically faster and most efficient
- **Inter-Rack communication** : Involves notes in different tracks and may have higher network latency and resource requirement due to the physical separation of the rex within the data center |

**Q. What is the responsibility of resource manager**
** **
**A.** It manages the resource within the cluster, It has two components within it called **application manager** and **scheduler** .
- **application manager**: It maintenance registry of running application and monitor their execution, application manager check client authentication.
- **scheduler**: Responsible for allocating cluster resource among application
  
Application managers responsibility is to accept job submissions negotiating the first container for executing the application a specific application master and provide the service for restarting the application master container on failure 

**Q. What is the data locality**
** **
**A.**  transfer code from application master to all node manager it is called data locality

**Important points**
- Resorts manager failure handled by another resource manager if it is created before application starting
- In case of application master failure, resource manager handles this problem and again create a new application masters
- In the case of node manager failure application will run another namenode and application master again allocate this task another namenode
- Must be no **application master** and **application manager** are different
- Application master runs on datanode
- Application manager runs on namenode
