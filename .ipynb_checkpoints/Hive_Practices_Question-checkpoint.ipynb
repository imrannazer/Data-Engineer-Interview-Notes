**Q. Explain all 4 types of table in Hive**
****
**A.**                   
**1. INTERNAL/MANAGED TABLE**: By default, Hive creates this table. Hive owns the data/file on the table meaning any data you insert or load files to the table are managed by the Hive process. **When you delete that table data and metadata both get deleted**.

`CREATE TABLE emp_table(id int, name string, age int, salary double)`   
`ROW FORMAT DELIMITED`             
`FIELDS TERMINATED BY ','`             
`STORED AS ORC`                  
`LOCATION '/hive/user/warehouse/';`              

****
**2. EXTERNAL TABLE**: Using "EXTERNAL" keyword you can create an external table. In external table "TRUNCATE" Table Command not supported. When you go to drop external table only table metadata from metastore will be removed but the underlying files will not be removed.

`CREATE EXTERNAL TABLE emp_table(id int, name string, age int, salary double)`   
`ROW FORMAT DELIMITED`             
`FIELDS TERMINATED BY ','`             
`STORED AS ORC`                  
`TBLPROPERTIES('key1 = value1', key2 = value2)`

****
**3. TEMPORARY TABLE**: It is created by using "TEMPORARY" keyword this table exists only within the current session (until hive is running).
- Cannot create partition table
- Indexes are not supported
- switch to us at users scratch directory "/tem/hive/<user>/"
- Automatically removed when session end
- It can be managed or external table

`CREATE TEMPORARY TABLE emp_table(id int, name string, age int, salary double)`   
`ROW FORMAT DELIMITED`             
`FIELDS TERMINATED BY ','`             
`STORED AS ORC`                  
`LOCATION '/hive/user/warehouse/';`              


****
**4. TRANSITIONAL TABLE**: Still transitional table DML(Insert, Update ,delete) command can be used. But the table should be **transactional(true)** table along with **orc** file format and should be **bucketed Table**

`> create table HiveTest1 (EmployeeID Int,FirstName String,Designation String, Salary Int,Department String)`         
`> clustered by (department) into 3 buckets`              
`> stored as orc`                  
`> TBLPROPERTIES ('transactional'='true');`

**Q. List of join types**

- **Inner Join**                  
  ` hive> select  e1.empname, e2.department_name  `                         
  ` from employee e1 `                 
  ` join employee_department e2 `                    
  ` on e1.empid= e2.depid; `                      

- **Left Outer Join**                  
  ` hive> select  e1.empname, e2.department_name  `                         
  ` from employee e1 `                 
  ` left outer join employee_department e2 `                    
  ` on e1.empid= e2.depid; `                      
- **Right Outer Join**                  
  ` hive> select  e1.empname, e2.department_name  `                         
  ` from employee e1 `                 
  ` right outer join employee_department e2 `                    
  ` on e1.empid= e2.depid; `                      
- **Full Outer Join**                  
  ` hive> select  e1.empname, e2.department_name  `                         
  ` from employee e1 `                 
  ` full outer join employee_department e2 `                    
  ` on e1.empid= e2.depid; `                     
- **left semi join**                  
  ` hive> select  e1.empname, e2.department_name  `                         
  ` from employee e1 `                 
  ` left semi employee_department e2 `                    
  ` on e1.empid= e2.depid; `

**Q. What is the skew join**
***
**A.** When there is a table which is skewed data(one is small and another is large) in the joining column we use skew join features              
- There is some properties

<table>
  <thead>
    <tr>
      <th><strong>Join Type</strong></th>
      <th>Skew Join</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><strong>When</strong></td>
      <td>Sorted the same</td>
    </tr>
    <tr>
      <td><strong>How</strong></td>
      <td>If a small number of skewed keys make up a significant percentage of the data, they will not become bottlenecks</td>
    </tr>
    <tr>
      <td><strong>Pointers</strong></td>
      <td>Because of partial results, the results also have to be read and written twice<br>The user needs to be aware of the skew in the data and manually do the above process</td>

- Set Hive properties for skew-join                             
  `SET hive.optimize.skewjoin=true;`                           
  `SET hive.skewjoin.key=join_column1,join_column2;`                             
  `SET hive.skewjoin.mapjoin.map.tasks=100;`                  
  `SET hive.skewjoin.mapjoin.min.split=256000000; -- 256 MB`                 
  `SET hive.skewjoin.mapjoin.hybrid.grace.factor=1.5;`                  
  `SET hive.skewjoin.reduce.memusage=0.1; -- 10%`
